```python
def get_address(df, index, one_or_zero):
    lon = df.iloc[index][0]
    lat = df.iloc[index][1]
    web_input = {'lon':str(lon), 
                  'lat':str(lat), 
                  'state':"or", 
                  'apikey':"06b0930203314ecc90082f30a9e65753", 
                 'format':"json",
                 'notStore':"false",
                 'version':"4.10"}
    website = requests.get('https://geoservices.tamu.edu/Services/ReverseGeocoding/WebService/v04_01/Rest/?',
                           params=web_input)
    json_read = json.loads(website.content)
    idx_cp = index
    if one_or_zero == 1:
        while json_read['QueryStatusCode'] != "Success":
            idx_cp = idx_cp + 1
            lon = df.iloc[idx_cp][0]
            lat = df.iloc[idx_cp][1]
            web_input = {'lon':str(lon), 
                      'lat':str(lat), 
                      'state':"or", 
                      'apikey':"06b0930203314ecc90082f30a9e65753", 
                     'format':"json",
                     'notStore':"false",
                     'version':"4.10"}
            website = requests.get('https://geoservices.tamu.edu/Services/ReverseGeocoding/WebService/v04_01/Rest/?',
                                   params=web_input)
            json_read = json.loads(website.content)
    else:
        while json_read['QueryStatusCode'] != "Success":
            idx_cp = idx_cp - 1 
            lon = df.iloc[idx_cp][0]
            lat = df.iloc[idx_cp][1]
            web_input = {'lon':str(lon), 
                      'lat':str(lat), 
                      'state':"or", 
                      'apikey':"06b0930203314ecc90082f30a9e65753", 
                     'format':"json",
                     'notStore':"false",
                     'version':"4.10"}
            website = requests.get('https://geoservices.tamu.edu/Services/ReverseGeocoding/WebService/v04_01/Rest/?',
                                   params=web_input)
            json_read = json.loads(website.content)
    print(json_read)
    address = str(json_read['StreetAddresses'][0]['StreetAddress'])
    df.iloc[index]['streetName'] = address
    return address
```

```python
turn_pts = list()
def binarySearch(df):
    print()
    print("start a new bs")
    # df = df.reset_index(drop=True)
    start = df.index[0]
    end = df.index[-1]
    mid = math.ceil((start + end) / 2)
    print(start, mid, end)
    # find start point address 
    start_address = get_address(df, start, 1)
    
    # find mid point address 
    mid_address = get_address(df, mid, 1)
    
    # find end point address 
    end_address = get_address(df, end, 0)
    
    # base case
    if len(df) <= 2:
        for index, item in df.iterrows():
            row, col = item
            l = [row, col]
            turn_pts.append(l)
        return None
    
    elif start_address == mid_address and mid_address == end_address:
        if end - start < math.ceil(len(df) / 1000):  # 
            return None
        else:
            binarySearch(df.iloc[start:mid])
            binarySearch(df.iloc[mid:end])
            
    else:
        binarySearch(df.iloc[start:mid])
        binarySearch(df.iloc[mid:end])
    print(turn_pts)
```
